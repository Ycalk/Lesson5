# Задание 1: Стандартные аугментации torchvision

```python
import utils
import torchvision.transforms as transforms
from PIL import Image
```

```python
dataset = utils.CustomImageDataset("data/train", transform=None, target_size=(224, 224))
```

```python
horizontal_flip = transforms.RandomHorizontalFlip(p=1)
vertical_flip = transforms.RandomVerticalFlip(p=1)
crop = transforms.RandomCrop(size=(224, 224), padding=50)
color_jitter = transforms.ColorJitter(
    brightness=0.3, contrast=0.3, saturation=0.3, hue=0.3
)
rotate = transforms.RandomRotation(degrees=30)
gray_scale = transforms.Grayscale(num_output_channels=3)
all_transforms = transforms.Compose(
    [
        horizontal_flip,
        vertical_flip,
        crop,
        color_jitter,
        rotate,
        gray_scale,
    ]
)


def show_augmentations(image: Image.Image, label: int):
    """Показывает оригинальное изображение и его аугментации.

    Args:
        image (Image.Image): Изображение для аугментации.
        label (int): Метка класса изображения, для отображения названия класса.
    """
    utils.show_images(
        [
            image,
            horizontal_flip(image),
            vertical_flip(image),
            crop(image),
            color_jitter(image),
            rotate(image),
            gray_scale(image),
            all_transforms(image),
        ],
        labels=[
            "Оригинал",
            "RandomHorizontalFlip",
            "RandomVerticalFlip",
            "RandomCrop",
            "ColorJitter",
            "RandomRotation",
            "Grayscale",
            "Все вместе",
        ],
        images_in_row=3,
        size=150,
        title=dataset.get_class_names()[label],
    )
```

### Гароу

```python
show_augmentations(*dataset[1])
```
